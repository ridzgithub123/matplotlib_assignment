Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
Note: Also add title, xlabel, and ylabel to the plot.

```python
'''
Scatter plots are the graphs that present the relationship between two variables in a data-set.
It represents data points on a two-dimensional plane or on a Cartesian system.
The independent variable or attribute is plotted on the X-axis, while the dependent variable is plotted on the Y-axis.
'''
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
plt.scatter(x,y)
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("random data scatter plot")
```

Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

```python
'''
A bar plot or bar chart is a graph that represents the category of data with rectangular bars with lengths and heights that is proportional to the values which they represent. 
The bar plots can be plotted horizontally or vertically. A bar chart describes the comparisons between the discrete categories.
One of the axis of the plot represents the specific categories being compared, while the other axis represents the measured values corresponding to those categories.
'''
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.bar(company,profit)
plt.show()
plt.barh(company,profit)
plt.show()
```

Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

```python
'''
 It is a type of chart that depicts a group of numerical data through their quartiles.
 It is a simple way to visualize the shape of our data. It makes comparing characteristics of data between categories very easy.
 
 Uses of a Box Plot

1) Box plots provide a visual summary of the data with which we can quickly identify the average value of the data, how dispersed the data is,
   whether the data is skewed or not (skewness).
2) The Median gives you the average value of the data.
3) Box Plots shows Skewness of the data-
a) If the Median is at the center of the Box and the whiskers are almost the 
   same on both the ends then the data is Normally Distributed.
b) If the Median lies closer to the First Quartile and if the whisker at the lower
   end is shorter (as in the above example) then it has a Positive Skew (Right Skew).
c) If the Median lies closer to the Third Quartile and if the whisker at the
   upper end is shorter then it has a Negative Skew (Left Skew).
4) The dispersion or spread of data can be visualized by the minimum and maximum values which are found at the end of the whiskers.
5) The Box plot gives us the idea of about the Outliers which are the points which are numerically distant from the rest of the data.

'''
box1 = np.random.normal(100, 10, 200)
plt.box(box1)
plt.show()
box2 = np.random.normal(90, 20, 200)
plt.box(box2)
plt.show()
```
